# Matplotlib:

*  Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
*  Matplotlib was created by John D. Hunter
*  Matplotlib is open source and we can use it freely
*  Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib.pyplot as plt
```

## 1.Line Chart:

* A line chart in Matplotlib is a graphical representation of data where points are plotted on a graph and connected by straight lines. 
* It is used to visualize trends, patterns, or relationships between variables, often over time.

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y)

plt.title("Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

## 2.Bar Chart:

### i.Single bar chart:

* A single bar plot in Matplotlib is a simple type of bar chart used to represent data as rectangular bars.
* It is used for comparing a single dataset across different categories.

```python
import numpy as np

fruits = ['Apples', 'Bananas', 'Fig', 'Mango']
sales = [400, 350, 300, 450]

plt.bar(fruits, sales,color='purple',edgecolor='black')
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

### ii.Multiple bar chart:

* A multiple bar plot (also called a grouped bar plot) in Matplotlib is a bar chart that displays multiple sets of data side by side for each category. 
* This type of plot is useful for comparing the values of multiple datasets within the same categories.


```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values_group1 = [10, 15, 20, 25]
values_group2 = [12, 18, 22, 28]

x = np.arange(len(categories)) 
bar_width = 0.4

plt.bar(x - bar_width/2, values_group1, width=bar_width, label='Group 1', color='blue',edgecolor='black')
plt.bar(x + bar_width/2, values_group2, width=bar_width, label='Group 2', color='orange',edgecolor='black')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Multi-Bar Plot')
plt.xticks(x, categories)  # Set the x-axis labels to the category names
plt.legend()

plt.show()
```

## 3.Histogram:

* A histogram in Matplotlib is a graphical representation of the distribution of numerical data, created using the `plt.hist()`function.
* It divides data into bins (intervals) and shows the frequency of data points within each bin.

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

plt.show()
```

## 4.Scatter Plot:

* A scatter plot in Matplotlib is a graph used to visualize the relationship between two continuous variables by plotting points based on their x and y values. 
* It is created using the `plt.scatter(x, y)` function.
* where `x` and `y` are the coordinates of the points.

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x, y)

plt.legend("A")

plt.title("Scatter chart")
plt.show()
```

## 5.Pie Chart:

* A pie chart in Matplotlib is a circular graph used to display proportional data, with each slice representing a fraction of the whole.
* It is created using the `plt.pie()` function,
* which allows customization of colors, labels, and percentages.

```python
x = [1, 2, 3, 4] 

e  =(0.1, 0, 0, 0)

plt.pie(x, explode = e )

plt.title("Pie chart")
plt.show()
```

## 6. Stack Plot:

* A stack plot in Matplotlib is a type of area chart that displays multiple datasets stacked on top of each other.
* showing how individual components contribute to the whole over a continuous variable (like time). 
* It is created using the `plt.stackplot()` function.

```python
days = [1, 2, 3, 4, 5]
Studying = [7, 8, 6, 11, 7]
playing = [8, 5, 7, 8, 13]

plt.stackplot(days, Studying, playing, colors =['r', 'c'])
plt.xlabel('Days')

plt.ylabel('No of Hours')

plt.title('Representation of Study and \
Playing wrt to Days')

plt.show()
```

## 7.Stem Plot:

* A stem plot in Matplotlib is a graph that displays data as vertical lines (stems) originating from a baseline, with markers at the ends, representing discrete data points. 
* It is commonly used to visualize sequences or time series data.

```python
plt.stem(x, y)
plt.show()
```

## 8.Box Plot:

* A box plot in Matplotlib is a statistical plot that displays the distribution of data through its quartiles, highlighting the median, interquartile range (IQR), and potential outliers.
* It is created using the plt.boxplot() function, ideal for visualizing the spread and variability of numerical datasets.

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(8, 6))

plt.boxplot(data)

plt.show()
```

## 9.Error Plot:

* An error plot in Matplotlib is a graphical representation of data with associated error bars, indicating variability or uncertainty in the measurements.
* It is created using the plt.errorbar() function, which adds vertical or horizontal error bars to the data points.

```python
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]

y_error = 0.2

plt.plot(x, y)

plt.errorbar(x, y,yerr = y_error,fmt ='o')
```

## 10.Violin Plot:

* A violin plot in Matplotlib combines aspects of a box plot and a kernel density plot to visualize the distribution of data, showing the probability density at different values.  
* It displays data distribution with mirrored density plots on each side, providing more insight into the shape and spread of the data.
